```python
# BASIC QUIRES
# WHICH TEAM WON MOST MATCHES
match_winners=df['match_winner'].value_counts()
sns.barplot(y=match_winners.index, x=match_winners.values, palette='viridis')
plt.title("MOST MATCH WIN BY A TEAM ")
```

```python
# TOSS DECISION TRENDS
sns.countplot(x=df['toss_decision'], palette='viridis')
plt.title('TOSS DECISIONS')
```

```python
# TOSS WINNERS VS MATCH WINNERS.
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percentage=(count*100)/df.shape[0]
percentage
```

```python
# HOW TEAMS WON (RUNS VS WICKETS)
sns.countplot (x=df['won_by'])
plt.title("WON BY")
```

```python
# PLAYER PERFORMAENCE
# Top 10 players with most Player of the Match awards
count = df['player_of_the_match'].value_counts().head(10)

# Plot
sns.barplot(x=count.values, y=count.index, palette='coolwarm')

# title
plt.title('PLAYER PERFORMANCES ')
```

```python
#  TWO TOP SCORERS
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high.plot(kind='barh')
plt.title("TOP SCORERS")
```

```python
# Extract wickets (if not already correct)
df['highest_wicket'] = df['best_bowling_figure'].astype(str).apply(lambda x: x.split('-')[0])
df['highest_wicket'] = df['highest_wicket'].astype(int)

# Group by correct column
top_bowlers = df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending=False).head(10)

# Top 10
# top_bowlers.head(10)

top_bowlers.plot(kind='barh')
```

```python
# VENUE ANALYSIS
venue_count = df['venue'].value_counts()
sns.barplot(x=venue_count.values, y=venue_count.index, palette='coolwarm')
```

```python
# COSTOM QUIRES
# WHO WON BY HIGHEST MARGIN
df[df['won_by'] == 'Runs'].sort_values(by = 'margin',ascending=False).head(1)[['match_winner','margin']]
```

```python
# COSTOM QUIRES
# WHICH PLAYER HAD THE HIGHEST INDIVIDUAL SCORE
df[df['highscore'] == df['highscore'].max()][["top_scorer",'highscore']]
```

```python
# COSTOM QUIRES
# WHICH PLAYER HAD THE BEST BOWLING FIGURE
df[df['highest_wicket'] == df ['highest_wicket'].max()]
```
